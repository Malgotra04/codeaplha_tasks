```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
import nltk
import warnings

warnings.filterwarnings('ignore')
nltk.download('punkt', quiet=True)

data = {
    'Review': [
        'This product is amazing and works great!',
        'Terrible experience, waste of money.',
        'It’s okay, not too bad but not the best.',
        'Absolutely fantastic quality and fast delivery!',
        'Poor packaging but decent product.'
    ]
}

df = pd.DataFrame(data)

def get_sentiment(review):
    analysis = TextBlob(review)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['Review'].apply(get_sentiment)

sentiment_counts = df['Sentiment'].value_counts()
positive_percent = (sentiment_counts.get('Positive', 0) / len(df)) * 100
negative_percent = (sentiment_counts.get('Negative', 0) / len(df)) * 100
neutral_percent = (sentiment_counts.get('Neutral', 0) / len(df)) * 100

print(" Sentiment Analysis Completed Successfully!\n")
print(df)
print("\n Insights:")
print(f"Positive Reviews: {positive_percent:.1f}%")
print(f"Negative Reviews: {negative_percent:.1f}%")
print(f"Neutral Reviews: {neutral_percent:.1f}%")

print("\n Business Insights:")
print("- Focus marketing around features that customers praise in positive reviews.")
print("- Investigate common complaints in negative reviews to improve product experience.")
print("- Use neutral feedback to identify areas for improvement without major issues.\n")

plt.figure(figsize=(7,5))
plt.bar(sentiment_counts.index, sentiment_counts.values, color=['green', 'red', 'gold'])
plt.title("Sentiment Distribution in Amazon Reviews", fontsize=13, fontweight='bold')
plt.xlabel("Type of Reviews")
plt.ylabel("Number of Reviews")
plt.show()
```
